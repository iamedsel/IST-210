My ER Diagram consists of six tables that are all interconnected to most. My six tables include Pet, Pet Food, Pet Transaction, Enclosure, Loyalty Customer, and Pet Food Transaction. In the pet table, "KindOfFood" is associated with "KindOfFood" in the Pet Food table. This association comes from the species of pet eating their food type. Enclosure ID is associated with species, due to only a specific species living with the same species, in one enclosure. In the Loyalty Customer table, "customerID" is an attribute also in "Pet Food Transaction" and "Pet Transaction". The reason for the customerID relation is when an item is purchased, it is saved to you customerID if the customer has an account. Food price is also shared in the Pet Food table and Pet Food Transaction table. The association behind this is because the price of food does not change between the two tables.
