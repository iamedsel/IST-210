Description:

Using their PSU IDs, the customer can use the access portal to get onto the website in order to begin the process of using our store. This is a 1 to 1 relationship for security purposes, as you wouldn't want someone else logging on as you, and using your information. After this, a customer can search for, filter, and purchase products. A customer can purchase multiple products at a time or no products at all, which is why this one is a one to many relationship. Once a product is purchased, the shipping company receives the product or products bought by the customer, and sends it to the current address of the customer. Should there be any problems, the support team moderates shipping, and any issue can hopefully be resolved by giving the order number or tracking number to a representative who can see where the package is at that given time.
